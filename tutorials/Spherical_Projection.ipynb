````python
from Cloud2DImageConverter import api
````

The **run** function has the following structure;
```python
def run(data_path, results_path, batch_size=500, fov_up=3.0, fov_down=-25.0, width=1024, height=64, is_label=True):
```
Where: 
* data_path: the path containing the velodyne and labels folders. 
* results_path: the path where the results will be saved. 
* batch_size: the amount of data that will be loaded into memory in each iteration. 
* fov_up, fov_down, width, and height: the parameters of the projection. 
* is_label: a boolean value that indicates whether the data contains the label folder. 

````python
api.run(data_path="../point_clouds/semantic_kitti/", results_path="../results/")
````
